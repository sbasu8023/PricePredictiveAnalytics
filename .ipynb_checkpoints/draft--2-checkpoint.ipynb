```python
# Removing Column Name Space 
def removeColmnSapce(df):
    coldict={}
    for i in df.columns:
        coldict[i]=i.replace(' ', '')
    
    return df.rename(columns=coldict)
train=removeColmnSapce(train)
test=removeColmnSapce(test)
```
